```python
def plot_pie_chart(column_name):
    data = df[column_name].value_counts()
    labels = data.index
    sizes = data.values

    # Define colors using an aesthetic colormap
    colors = plt.cm.Set3.colors  # Soft pastel colors for aesthetics
    explode = [0.05] * len(labels)  # Slightly separate each slice for a 3D effect

    plt.figure(figsize=(8, 8))
    plt.pie(
        sizes, labels=labels, autopct='%1.1f%%', startangle=90,
        colors=colors[:len(labels)], explode=explode, shadow=True
    )
    plt.axis('equal')  # Ensures pie chart is a circle
    plt.title(f'Percentage of {column_name}', fontsize=16, color="#4A4A4A", fontweight='bold')
    plt.show()

# Plot each categorical column with enhanced styling
plot_pie_chart('gender')
plot_pie_chart('race/ethnicity')
plot_pie_chart('lunch')
plot_pie_chart('parental level of education')
plot_pie_chart('test preparation course')
```

```python
def plot_pie_chart(column_name):
    data = df[column_name].value_counts()
    labels = data.index
    sizes = data.values

    # Use pastel colors by sampling from Pastel1 colormap
    cmap = plt.cm.Pastel1
    colors = [cmap(i / len(labels)) for i in range(len(labels))]

    plt.figure(figsize=(7, 7))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
    plt.axis('equal')  # Ensures pie chart is a circle
    plt.title(f'Percentage of {column_name}', fontsize=16, color="#4A4A4A", fontweight='bold')
    plt.show()

# Plot each categorical column with pastel colors
plot_pie_chart('gender')
plot_pie_chart('race/ethnicity')
plot_pie_chart('lunch')
plot_pie_chart('parental level of education')
plot_pie_chart('test preparation course')
```

Prompt:

Beri kode python untuk melihat jumlah siswa yang fail dan pass untuk masing-masing pelajaran (math, reading, writing score). Dengan minimal nilai yang dianggap pass adalah 75!

```python
# Define the passing score
passing_score = 75

# Create pass/fail status columns for each subject
df['math_status'] = df['math score'].apply(lambda x: 'Pass' if x >= passing_score else 'Fail')
df['reading_status'] = df['reading score'].apply(lambda x: 'Pass' if x >= passing_score else 'Fail')
df['writing_status'] = df['writing score'].apply(lambda x: 'Pass' if x >= passing_score else 'Fail')

# Set up the figure and size
plt.figure(figsize=(15, 5))

# Math Score Distribution
plt.subplot(1, 3, 1)
sns.boxplot(data=df, x='math_status', y='math score', palette='coolwarm')
plt.title('Math Score Distribution')
plt.xlabel('Status')
plt.ylabel('Math Score')

# Reading Score Distribution
plt.subplot(1, 3, 2)
sns.boxplot(data=df, x='reading_status', y='reading score', palette='coolwarm')
plt.title('Reading Score Distribution')
plt.xlabel('Status')
plt.ylabel('Reading Score')

# Writing Score Distribution
plt.subplot(1, 3, 3)
sns.boxplot(data=df, x='writing_status', y='writing score', palette='coolwarm')
plt.title('Writing Score Distribution')
plt.xlabel('Status')
plt.ylabel('Writing Score')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

Prompt: Beri kode untuk melihat nilai rata-rata seluruh mata pelajaran (math, reading writing) dari setiap siswa dan lihat korelasinya dengan masing-masing kategori

```python
# Menggunakan dataset contoh
data = {
    'gender': ['female', 'male', 'female', 'male', 'female', 'male', 'female', 'male'],
    'race/ethnicity': ['group A', 'group B', 'group C', 'group D', 'group E', 'group A', 'group B', 'group C'],
    'parental level of education': ['high school', 'some college', 'associate degree', 'bachelor\'s degree', 'high school', 'some college', 'associate degree', 'bachelor\'s degree'],
    'lunch': ['standard', 'free/reduced', 'standard', 'free/reduced', 'standard', 'free/reduced', 'standard', 'free/reduced'],
    'test preparation course': ['completed', 'none', 'completed', 'none', 'completed', 'none', 'completed', 'none'],
    'math score': [72, 65, 90, 88, 78, 85, 95, 82],
    'reading score': [74, 63, 92, 85, 76, 88, 94, 81],
    'writing score': [78, 65, 85, 89, 77, 90, 92, 80],
}

# Membuat DataFrame
df = pd.DataFrame(data)

# Menghitung rata-rata nilai untuk setiap siswa
df['average score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

# Mengatur ukuran plot
plt.figure(figsize=(12, 8))

# Visualisasi rata-rata nilai dengan kategori gender
plt.subplot(2, 2, 1)
sns.boxplot(x='gender', y='average score', data=df)
plt.title('Average Score by Gender')
plt.ylabel('Average Score')

# Visualisasi rata-rata nilai dengan kategori ras/etnis
plt.subplot(2, 2, 2)
sns.boxplot(x='race/ethnicity', y='average score', data=df)
plt.title('Average Score by Race/Ethnicity')
plt.ylabel('Average Score')

# Visualisasi rata-rata nilai dengan tingkat pendidikan orang tua
plt.subplot(2, 2, 3)
sns.boxplot(x='parental level of education', y='average score', data=df)
plt.title('Average Score by Parental Level of Education')
plt.ylabel('Average Score')
plt.xticks(rotation=45)

# Visualisasi rata-rata nilai dengan jenis makan siang
plt.subplot(2, 2, 4)
sns.boxplot(x='lunch', y='average score', data=df)
plt.title('Average Score by Lunch Type')
plt.ylabel('Average Score')

# Menampilkan plot
plt.tight_layout()
plt.show()

# Menampilkan DataFrame dengan nilai rata-rata
print(df[['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course', 'average score']])
```
